```python
import pandas as pd
import matplotlib.pyplot as plt

   
     
df=pd.read_csv("E:\\veteran-suicides.csv")   
#print(df.head())
```

```python
# index set to state column

df2 = df.set_index("state", drop = False)
rows=df['vetsuicides']

 
```

```python
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print("Current size:", fig_size)
 
```

```python
# Set figure width to 12 and height to 9

fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size
```

```python
# veteran suicides in each state of USA

plt.hist(rows,bins=5)
plt.title('number of suicides in each state of USA ')

# Analyzing veteran suicides in the year 2011 state wise, USA 

# Three states show more number of suicides in a range of 500-800
# Rest 47 states are below 300
```

```python
# Veteran Suicide rate in each state of USA

plt.hist(df['vetrate'],bins=5)
plt.title('veteran suicide rate in 2011 USA')  

# Analyzing which state has high veteran suicide 

# three states (different from the above graph ) show high veteran suicide rate 
# Around 18 countries show (35-45)% of suicidal rate
```

```python
# line graph representation of suicidal rate based on gender
   
fig, ax1 = plt.subplots()




a=df2['vetmalesp']
b=df2['vetfemalesp']
 
color = 'tab:red'
ax1.set_xlabel('states')
# to rotate the values of x-axis  
plt.xticks(rotation=90)
ax1.set_ylabel('malep', color=color)
ax1.plot(a, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('femalep', color=color)  # we already handled the x-label with ax1
ax2.plot(b,color=color)


# to set dual axes to the same scale 

ax2.set_ylim(0,1)

ax2.tick_params(axis='y', labelcolor=color)
plt.title('suicidal rate based on  gender')

# Analyzing which gender has high veteran suicidal rate

# The below graph depicts that male has high suicidal rate than female
# Highest Male suicidal rate: Tennessee, Arkansas
# Lowest Male suicidal rate: South Dakota, Arizona, Illinois, lowa, Oklahoma, Rhode island, Wisconsin, Wyoming
# Highest female suicidal rate: Delaware and Alaska 
# Lowest female suicidal rate: Arizona, Georgia, Wisconsin, Wyoming, louisiana, Georgia, Nebraska, Newada
```

```python
# stacked bar graph representation of suicidal rate based on gender

r=df['state']

p1 = plt.bar(r,a,bottom=b,color='tab:red')
p2 = plt.bar(r,b,color='tab:blue')


plt.ylabel('Male and Female ratio')
# to rotate the values of x-axis  
plt.xticks(r,rotation=90)

plt.legend((p1[0], p2[0]), ('Women', 'Men'))


fig.tight_layout()


plt.show()

# Suicidal rate based on gender (Stacked bar representation)
```

```python
# Analysis on the suicidal rate based on the age groups

r=df['state']

age1=df['age15to34']
age2=df['age35to44']
age3=df['age45to59']
age4=df['age60to65']

p1 = plt.bar(r,age1,bottom=age2,color='tab:red')
p2 = plt.bar(r,age2,bottom=age3,color='tab:blue')
p3 = plt.bar(r,age3,bottom=age4,color='tab:green')
p4 = plt.bar(r,age4,color='tab:purple')




plt.ylabel('suicides')
# to rotate the values of x-axis  

plt.xticks(r,rotation=90)


plt.legend((p1[0], p2[0],p3[0],p4[0]), ('15to34', '35to44','45to59','60to65'))





plt.show()

# Below graph depicts suicidal rate among different age groups 
# Graph Analysis: Suicidal rate is high in the age group of 60-65
# High rate: California
# Low rate: Alabama, Conneticut, Florida 
```
